# Classification workshop

<b>Classification algorithms</b> are useful in extracting meaning from large sets of incoming data. 

The process of training a classifier follows:
1. Organise and clean the data
2. Divide the data into training and testing subset
3. Use the classifier to associated feature attributes within the training data to known classifications
4. Test the strength of the model fit by predictin classifications within the test dataset
5. Apply the classifier to any new and future data to establish classifications



```python
import pandas as pd
```

```python
df = pd.read_csv("Data/iris.csv")
```

```python
df.tail(5)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.drop(columns = "Id", inplace =True)
```

```python
df.Species.value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit(["Iris-setosa", "Iris-versicolor", "Iris-virginica"])
df.Species = le.transform(df.Species)
```

```python
df.drop(columns = "Id", inplace = True)
df
```

```python
x = df.drop(columns = "Species")
y = df["Species"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.3)
```

```python
x_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
clf_knn = KNeighborsClassifier(n_neighbors=60)
clf_knn.fit(x_train, y_train)
```

```python
clf_knn.score(x_train, y_train)
```

```python
test_pred_knn = clf_knn.predict(x_test)
```

```python
from sklearn import metrics
confusion_matrix = metrics.confusion_matrix(y_test, test_pred_knn)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.matshow(confusion_matrix)
plt.title('confusion matrix')
plt.colorbar()
plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.show()
```

```python
print(metrics.classification_report(y_test, test_pred_knn))
```

```python
from sklearn.model_selection import GridSearchCV
# Set the parameters by cross-validation

tuned_parameters = [{'n_neighbors': [5,10, 20, 50, 80]}]


scores = ['precision', 'recall']
for score in scores:
    print("# Tuning hyper-parameters for %s" % score)
    print()

    clf = GridSearchCV(
        KNeighborsClassifier(), tuned_parameters, scoring='%s_macro' % score
    )
    clf.fit(x_train, y_train)

    print("Best parameters set found on development set:")
    print()
    print(clf.best_params_)
    print()
    print("Grid scores on development set:")
    print()
    means = clf.cv_results_['mean_test_score']
    stds = clf.cv_results_['std_test_score']
    for mean, std, params in zip(means, stds, clf.cv_results_['params']):
        print("%0.3f (+/-%0.03f) for %r"
              % (mean, std * 2, params))
```

# Logistic regression

```python
from sklearn.linear_model import LogisticRegression

clf_logit_reg = LogisticRegression(random_state=0, max_iter=10e7).fit(x_train, y_train)
```

```python
test_pred_logit_reg = clf_logit_reg.predict(x_test)
print(metrics.classification_report(y_test, test_pred_logit_reg))
```

# Decision tree

```python
from sklearn.tree import DecisionTreeClassifier
clf_decision_tree = DecisionTreeClassifier()  # creates the kNN classifier, setting it to check the 60 neighbouring points
clf_decision_tree.fit(x_train, y_train)
test_pred_decision_tree = clf_decision_tree.predict(x_test)
```

```python
print (metrics.classification_report(y_test, test_pred_decision_tree))
```

```python
print(f'Model Accuracy: {clf_decision_tree.score(x_train, y_train)}')
```

```python
# the code is based on this link: https://stackoverflow.com/a/59517013/4667568
decision_tree_depth_5 = DecisionTreeClassifier(max_depth=3)
decision_tree_depth_5.fit(x_train, y_train)

plt.figure(figsize=(40,20))  # customize according to the size of your tree. If you can't see the tree clearly, you may change the size

from sklearn import tree
#start_time = time.time()

# vec.feature_names_ represents the variable names
#_ = tree.plot_tree(decision_tree_depth_5, feature_names = )
# if you want to save the picture to a file, add this line before plt.show()
#plt.savefig('decision_tree.png')
_ = tree.plot_tree(decision_tree_depth_5)

plt.show()
```

```python
from sklearn.tree import export_text
tree_rules = export_text(decision_tree_depth_5)
print(tree_rules)
```

```python
confusion_matrix = metrics.confusion_matrix(y_test ,test_pred_decision_tree)
```

```python
import seaborn as sns
# convert array to dataframe to plot
matrix_df = pd.DataFrame(confusion_matrix)

# set axis to add title and axis labels later
ax = plt.axes()
sns.set(font_scale=1.3) # for label size
plt.figure(figsize=(10,7))
cmap = sns.cubehelix_palette(8) # color palette
sns.heatmap(matrix_df, annot=True, fmt='g', ax=ax, cmap = cmap) #fmt so that numbers aren't scientific

#axis labels and title
ax.set_title('Confusion Matrix - Decision Tree', fontsize = 17)
ax.set_xlabel('Predicted Label', fontsize = 15)
ax.set_ylabel('True Label'     , fontsize = 15)
plt.show()
```

```python
metrics.accuracy_score(y_test, test_pred_decision_tree)
```

```python
precision = metrics.precision_score(y_test, test_pred_decision_tree, average = None)
precision_results = pd.DataFrame(precision)

precision_results.rename(columns={0:'Precision'}, inplace=True)
precision_results
```

```python
recall = metrics.recall_score(y_test, test_pred_decision_tree, average = None)
recall_results = pd.DataFrame(recall)

recall_results.rename(columns = {0:'Recall'}, inplace=True)
recall_results
```
